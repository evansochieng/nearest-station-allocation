```python
#load required libraries
import os
import pandas as pd
import numpy as np
from haversine import haversine
```

```python
#change your working directory
os.chdir("C:/Users/HP/Desktop/Syngenta/Jun-Aug 2021/Station Allocation/")
```

```python
#read in the required data
farmers = pd.read_csv("C:/Users/HP/Desktop/Syngenta/Jun-Aug 2021/Station Allocation/Farmers.csv")
stations = pd.read_csv("C:/Users/HP/Desktop/Syngenta/Jun-Aug 2021/Station Allocation/RG GPS.csv")
```

```python
#create an empty dataframe
col_names = list(stations.index)
col_names.insert(len(col_names)+1, "Closest")
station_allocation = pd.DataFrame("NA", index = farmers.index, columns = col_names)
station_allocation
```

```python
#calculate the distance of each python from the stations
farmers_latitude = farmers[["Latitude"]]
farmers_longitude = farmers[["Longitude"]]
farmers_coordinates = []
for i in range(len(farmers_latitude.index)):
    farmers_coordinates.insert(i, (float(farmers_latitude.loc[i].values), float(farmers_longitude.loc[i].values)))
stations_latitude = stations[["Latitude"]]
stations_longitude = stations[["Longitude"]]
for station in range(len(stations.index)):
    current_station = (float(stations_latitude.loc[station].values), float(stations_longitude.loc[station].values))
    for k in range(len(farmers_coordinates)):
        station_allocation.loc[k, station] = haversine(farmers_coordinates[k], current_station)
```

```python
#assign farmers closest stations
station = stations[["Name"]]
for farmer in range(len(station_allocation.index)):
    station_allocation.loc[farmer, "Closest"] = str(station.loc[np.array(station_allocation.iloc[farmer, 0:3]).argmin()].values)[2:-2]
```

```python
#Rename indexes and columns
col_names = list(stations["Name"])
col_names.insert(len(col_names)+1, "Closest")

station_allocation.columns = col_names
station_allocation.index = farmers["Farmer"]
```

```python
#save the dataframe with allocated stations
station_allocation.to_csv("Station_Allocated.csv")
```
